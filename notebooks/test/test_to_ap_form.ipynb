```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('../..')
import dizx
```

```python
zcircuit = dizx.Circuit(3)
zcircuit.add_gate("Z", 0)
zcircuit.add_gate("H", 1)
zcircuit.add_gate("NEG", 0)
zcircuit.add_gate("Z", 1)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("Z", 0)
zcircuit.add_gate("Z", 0)
zcircuit.add_gate("NEG", 1)
zcircuit.add_gate("NEG", 0)
zcircuit.add_gate("CZ", 0, 2)
zcircuit.add_gate("CZ", 2, 1)
zcircuit.add_gate("CZ", 0, 1)
zcircuit.add_gate("Z", 1)
zcircuit.add_gate("Z", 2)
zcircuit.add_gate("NEG", 2)
zcircuit.add_gate("NEG", 1)
g = zcircuit.to_graph(zh=True)
dizx.draw(g)
```

```python
from dizx.simplify import to_ap_form

to_ap_form(g)
dizx.draw(g)
```
